```python
golf_access_rule_metadata_df = pd.read_csv("datasets\metadata\golf_access_rule_meta_data.csv")
```

```python
%%time
conn = sqlite3.connect("golfBenefit.db")
cur = conn.cursor()
golf_access_rule_df.to_sql("golfAccessRule", conn, if_exists="replace")
golf_access_rule_metadata_df.to_sql("golfAccessRuleMetadata", conn, if_exists="replace")
```

```python
pd.read_sql_query("select * from golfAccessRule", conn)
```

```python
query = """select golfAccessRule.benefit_id,
         golfAccessRule.access_rule_id,
         golfAccessRule.access_rule,
         golfAccessRuleMetadata.Sunday_Is_Open,
         golfAccessRuleMetadata.Sunday_All_Day,
         golfAccessRuleMetadata.Sunday_Start_Time, 
         golfAccessRuleMetadata.Sunday_End_Time,
         golfAccessRuleMetadata.Monday_Is_Open,
         golfAccessRuleMetadata.Monday_All_Day,
         golfAccessRuleMetadata.Monday_Start_Time,
         golfAccessRuleMetadata.Monday_End_Time,
         golfAccessRuleMetadata.Tuesday_Is_Open,
         golfAccessRuleMetadata.Tuesday_All_Day,
         golfAccessRuleMetadata.Tuesday_Start_Time,
         golfAccessRuleMetadata.Tuesday_End_Time,
         golfAccessRuleMetadata.Wednesday_Is_Open,
         golfAccessRuleMetadata.Wednesday_All_Day,
         golfAccessRuleMetadata.Wednesday_Start_Time,
         golfAccessRuleMetadata.Wednesday_End_Time,
         golfAccessRuleMetadata.Thursday_Is_Open,
         golfAccessRuleMetadata.Thursday_All_Day,
         golfAccessRuleMetadata.Thursday_Start_Time,
         golfAccessRuleMetadata.Thursday_End_Time,
         golfAccessRuleMetadata.Friday_Is_Open,
         golfAccessRuleMetadata.Friday_All_Day,
         golfAccessRuleMetadata.Friday_Start_Time,
         golfAccessRuleMetadata.Friday_End_Time, 
         golfAccessRuleMetadata.Saturday_Is_Open,
         golfAccessRuleMetadata.Saturday_All_Day, 
         golfAccessRuleMetadata.Saturday_Start_Time,
         golfAccessRuleMetadata.Saturday_End_Time
         from golfAccessRule
         left join golfAccessRuleMetadata on trim(upper(golfAccessRuleMetadata.Availability)) = trim(upper(golfAccessRule.access_rule))"""

```

```python
rs_benefit = pd.read_sql_query(query, conn)
print("Total number of Access Rules: " + str(len(rs_benefit)))
```
