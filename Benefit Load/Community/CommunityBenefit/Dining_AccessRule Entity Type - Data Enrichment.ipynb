```python
%%time
conn = sqlite3.connect("diningBenefit.db")
cur = conn.cursor()
dining_access_rule_df.to_sql("diningAccessRule", conn, if_exists="replace")
dining_access_rule_metadata_df.to_sql("diningAccessRuleMetadata", conn, if_exists="replace")
```

```python
pd.read_sql_query("select * from diningAccessRule", conn)
```

```python
query = """select diningAccessRule.benefit_id,
         diningAccessRule.access_rule_id,
         diningAccessRule.access_rule,
         diningAccessRuleMetadata.Sunday_Is_Open,
         diningAccessRuleMetadata.Sunday_All_Day,
         diningAccessRuleMetadata.Sunday_Start_Time, 
         diningAccessRuleMetadata.Sunday_End_Time,
         diningAccessRuleMetadata.Monday_Is_Open,
         diningAccessRuleMetadata.Monday_All_Day,
         diningAccessRuleMetadata.Monday_Start_Time,
         diningAccessRuleMetadata.Monday_End_Time,
         diningAccessRuleMetadata.Tuesday_Is_Open,
         diningAccessRuleMetadata.Tuesday_All_Day,
         diningAccessRuleMetadata.Tuesday_Start_Time,
         diningAccessRuleMetadata.Tuesday_End_Time,
         diningAccessRuleMetadata.Wednesday_Is_Open,
         diningAccessRuleMetadata.Wednesday_All_Day,
         diningAccessRuleMetadata.Wednesday_Start_Time,
         diningAccessRuleMetadata.Wednesday_End_Time,
         diningAccessRuleMetadata.Thursday_Is_Open,
         diningAccessRuleMetadata.Thursday_All_Day,
         diningAccessRuleMetadata.Thursday_Start_Time,
         diningAccessRuleMetadata.Thursday_End_Time,
         diningAccessRuleMetadata.Friday_Is_Open,
         diningAccessRuleMetadata.Friday_All_Day,
         diningAccessRuleMetadata.Friday_Start_Time,
         diningAccessRuleMetadata.Friday_End_Time, 
         diningAccessRuleMetadata.Saturday_Is_Open,
         diningAccessRuleMetadata.Saturday_All_Day, 
         diningAccessRuleMetadata.Saturday_Start_Time,
         diningAccessRuleMetadata.Saturday_End_Time
         from diningAccessRule
         left join diningAccessRuleMetadata on trim(upper(diningAccessRuleMetadata.Availability)) = trim(upper(diningAccessRule.access_rule))"""
```

```python
rs_benefit = pd.read_sql_query(query, conn)
print("Total number of Access Rules: " + str(len(rs_benefit)))
```
